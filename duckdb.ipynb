## WORKING WITH TABLES

##### CREATING A TABLE

```python
# Create a table
conn.execute("""
    CREATE TABLE employees (
        id INTEGER,
        name VARCHAR,
        department VARCHAR,
        salary INTEGER,
        hire_date DATE
    )
""")

# Insert data
conn.execute("""
    INSERT INTO employees VALUES 
    (1, 'Alice', 'Engineering', 80000, '2020-01-15'),
    (2, 'Bob', 'Marketing', 65000, '2019-03-20'),
    (3, 'Charlie', 'Engineering', 90000, '2018-06-10'),
    (4, 'Diana', 'Sales', 70000, '2021-02-28'),
    (5, 'Eve', 'Marketing', 60000, '2022-07-15')
""")
```

```python
# Basic SELECT
result = conn.execute("SELECT * FROM employees").fetchall()
print("All employees:")
for row in result:
    print(row)
```

checked

```python
# SELECT with filtering
result = conn.execute("""
    SELECT name, department, salary 
    FROM employees 
    WHERE salary > 70000
""").fetchdf()  # Returns as pandas DataFrame

print("High earners:")
result
```

```python
# Create a DataFrame
df = pd.DataFrame({
    'product_id': [1, 2, 3, 4, 5],
    'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'price': [999.99, 29.99, 79.99, 299.99, 149.99],
    'category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Audio']
})

# Register DataFrame as a DuckDB table
conn.register('products', df)

# Query the DataFrame
result = conn.execute("""
    SELECT category, AVG(price) as avg_price
    FROM products
    GROUP BY category
""").fetchdf()

result
```

```python
# Query DataFrame directly without registering
result = duckdb.sql("""
    SELECT * 
    FROM df 
    WHERE price > 100
""").df()  # Returns as DataFrame

result
```
